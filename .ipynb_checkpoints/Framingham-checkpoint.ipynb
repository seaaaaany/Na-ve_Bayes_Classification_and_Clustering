### 2. Create 2 contingency tables, one with death and sex and another with death and educ

A contingency table is a tabular representation of categorical data . 

A contingency table usually shows frequencies for particular combinations of values of two discrete random variable s X and Y. 

Each cell in the table represents a mutually exclusive combination of X-Y values.

```python
# dependencies
import pandas as pd
import numpy as np
from sklearn.naive_bayes import MultinomialNB
import statsmodels.tools.tools as stattools
```

```python
t_ds=pd.crosstab(Framingham_training['Death'],Framingham_training['Sex'])
t_ds['Total']=t_ds.sum(axis=1)
t_ds.loc['Total']=t_ds.sum()
print(t_ds)

t_de=pd.crosstab(Framingham_training['Death'],Framingham_training['Educ'])
t_de['Total']=t_de.sum(axis=1)
t_de.loc['Total']=t_de.sum()
print(t_de)
```
